```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
# send get request to the webpage server to get the source code of the page.
```

```python
page=requests.get("https://www.ci.richmond.ca.us/1404/Major-Projects")
```

```python
page
```

```python
#page content
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
#scraping first heading
first_heading=soup.find('a',class_="navMainItem mainNavItem mainNavItem1")
```

```python
first_heading.text
```

```python
#scraping multiple titles
titles=[]
for i in soup.find_all('h3', class_="subhead2"):
    titles.append(i.text)
```

```python
print(titles)
```

```python
#scraping sub titles
sub_titles=[]
for i in soup.find_all('a', class_="Hyperlink"):
    sub_titles.append(i.text)
```

```python
print(sub_titles)
```

2. bakersfield https://www.bakersfieldcity.us/518/Projects-Programs

```python
# send get request to the webpage server to get the source code of the page.
```

```python
page=requests.get("https://www.bakersfieldcity.us/518/Projects-Programs")
```

```python
page
```

```python
#page content
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
#scraping multiple left hand side titles
titles=[]
for i in soup.find_all('a', class_="navMainItem secondaryNavItem cpComponent"):
    titles.append(i.text)
```

```python
print(titles)
```

```python
#scraping projects & programs
projects=[]
for i in soup.find_all('a', class_="megaMenuItem--link"):
    projects.append(i.text)
```

```python
print(projects)
```

```python
#scraping quick links
quick_links=[]
for i in soup.find_all('span', class_="cp-fancierButton-text--text1--line-0"):
    quick_links.append(i.text)
```

```python
print(quick_links)
```
